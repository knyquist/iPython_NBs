# Generate Fake Trace 

Import random library 

```python
import random
```

Seed the random number generator.

```python
random.seed()
```

Generate 10000 gaussian-distributed random numbers with $\mu=0$ and $\sigma=1$. 

```python
noise = []
numDataPts = 1000;
for n in range(numDataPts):
    noise.append(random.gauss(0, .5))
for i in range(numDataPts):
    noise[i]+=5
print(noise[0:5])
```

Plot histogram data as well as scatterplot.

```python
hist(noise, bins=10, histtype='step');
xlabel("random number");
ylabel("counts");
```

```python
plot(noise);
xlabel("data point");
ylabel("random number");
```

Generate list of exponentially distributed numbers, which will comprise the distribution of dwell durations.

```python
dwell = []
meanDur = 24
i = numDataPts #i will be the iterable
while i != 0:
    dwellDur = random.expovariate(1.0/meanDur)
    dwellDur = int(round(dwellDur))
    if i - dwellDur >= meanDur:
        dwell.append(dwellDur)
        i -= dwellDur;
    else:
        dwell.append(i)
        i = 0;
dwell[0:5]
```

Confirm that sum of all dwells is equal to the numDataPts

```python
print "sum of dwell durations: " + str(sum(dwell)) + " number of data points: " + str(numDataPts) 
```

Histogram dwell durations, confirm single exponential behavior

```python
hist(dwell,bins=5, histtype='step');
title("histogram of dwell durations")
xlabel('dwell duration');
ylabel('counts');
```

Generate stationary noise trace from the gaussian noise and the exponential dwell durations

```python
trace=list(noise)
fit=[5 for i in range(numDataPts)]
start=0
stepSize=0

for eachDwell in range(len(dwell)):
    dwellDur=dwell[eachDwell]
    for pts in range(start,dwellDur+start):
        trace[pts]=trace[pts]+stepSize
        fit[pts]=fit[pts]+stepSize
    start=dwellDur+start
    stepSize+=1
```

```python
plot(trace);
plot(fit,'r',lw=2);
title("example fake trace");
xlabel("time"); #xlim(1000,1125)
ylabel("signal"); #ylim(55,62.5)
```
